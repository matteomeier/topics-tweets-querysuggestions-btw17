#### Context

Newspaper articles are popular and diverse data sources for data scientists in various fields. To collect this kind of information, I built newspipe a scalable pipeline to crawl and extract data from news websites. Showing the power and flexibility of the pipeline, I collected articles from German-speaking publishers by providing the pipeline only with the URL of the website. This dataset is the result of the demonstration. I mainly crawled articles from 15 news sources over an extended period of time. The remaining 33 news sources were only test runs.
Content

    author : author name as string (may contain unrelevant information)
    published: publish date as unix timestamp (int value)
    title: article title as string
    summary: summary of the article (not cleaned)
    text: article text as string
    link: link to news article
    tags: tags of the article (optional)
