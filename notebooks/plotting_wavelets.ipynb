```python
# necessary import
import numpy as np
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
from scipy import signal
import pandas as pd
```

```python
df_scale = pd.DataFrame()
df_loc = pd.DataFrame()
```

```python
points = 1000
a = 10
b = 50
c = 100
vec1 = signal.ricker(points, a)
vec2 = signal.ricker(points, b)
vec3 = signal.ricker(points, c)
```

```python
df_loc['a'] = vec1[400:980] * 1000 + 100
df_loc['b'] = vec1[410:990] * 1000 + 100
df_loc['c'] = vec1[420:] * 1000 + 100
df_loc['d'] = np.random.randint(400, size=len(df_loc))
```

```python
df_scale_mittel = pd.DataFrame(data={'Wert': vec2, 'Skalierung': 'mittel'})
df_scale_groß = pd.DataFrame(data={'Wert': vec3, 'Skalierung': 'groß'})

df_scale = pd.DataFrame(data={'Wert': vec1, 'Skalierung': 'klein'})
df_scale = df_scale.append(df_scale_mittel)
df_scale = df_scale.append(df_scale_groß)
df_scale.reset_index(inplace=True)
df_scale.rename(columns={'index':'Index'}, inplace=True)
df_scale.head()
```

```python
fig = px.line(df_scale, x='Index', y='Wert', facet_col='Skalierung',
              template='simple_white', color_discrete_sequence=px.colors.qualitative.Antique)
fig.show()
```

```python
fig = go.Figure()
# Create and style traces

fig.add_trace(go.Scatter(x=np.arange(1, len(df_loc)), y=df_loc['d'], name='Hashtag Zeitreihe',
                         line = dict(color='rgb(217, 175, 107)', width=0.5, dash='dash')))

fig.add_trace(go.Scatter(x=np.arange(1, len(df_loc)), y=df_loc['c'], name='Wavelet',
                         line=dict(color='rgb(133, 92, 117)', width=3,
                                  dash='dot')))
fig.add_trace(go.Scatter(x=np.arange(1, len(df_loc)), y=df_loc['b'], name='Wavelet',
                         line=dict(color='rgb(133, 92, 117)', width=3,
                                  dash='dot')))
fig.add_trace(go.Scatter(x=np.arange(1, len(df_loc)), y=df_loc['a'], name='Wavelet',
                         line=dict(color='rgb(133, 92, 117)', width=3)))

fig.add_annotation(x=120,
                   y=390,
                   ax=60,
                   ay=390,
                   xref='x',
                   yref='y',
                   axref='x',
                   ayref='y',
                   #text='Wavelet wird verschoben',
                   showarrow=True,
                   arrowhead=2,
                   arrowsize=1,
                   arrowwidth=2
                  )

# Edit the layout
fig.update_layout(xaxis_title='Zeit',
                  yaxis_title='Häufigkeit', template='simple_white')

fig.show()
```
